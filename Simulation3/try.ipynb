```python
condlist = []
AdjList = []
for i in range(2,15):
    L, A_gt, s0,D = Setup(i)
    Samples, T = CreateNoisySamples(L,s0,0.001,1e-5,10)
    sk, dk = Resample(3,Samples,0.001)
    U,V = UVmatrix(T,sk,dk,3,D)
    Adj, Cond = Learn(U,V)
    Adj_le, A_gt, err = ErrCheck(Adj, A_gt)
    err
    AdjList.append([Adj_le,A_gt])
    condlist.append(Cond)
```

```python
for k in range():
    subplot(14,2,k+1)
    matshow((AdjList[k])[0])
    
```
